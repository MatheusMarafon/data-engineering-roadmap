Cenário 1: Multiplicação de colunas (Preço * Qtde)

1. Importando bibliotecas

```python
import numpy as np
import time
import random
```

2. Criando dados pesados

```python
lista_precos = [random.uniform(10.0, 500.00) for _ in range(1_000_000)]
lista_qtde = [random.randint(1, 10) for _ in range(1_000_000)]

np_precos = np.array(lista_precos)
np_qtde = np.array(lista_qtde)

print(f"Dados Gerados: {len(lista_precos)} linhas.")
```

3. O jeito 'lento' (loop for)

```python
%%time

faturamento_loop = []
for i in range(len(lista_precos)):
    calculo = lista_precos[i] * lista_qtde[i]
    faturamento_loop.append(calculo)
```

4. O jeito 'rápido' (vetorização)

```python
%%time

faturamento_numpy = np_precos * np_qtde
```

Cenário 2: Filtragem condicional (Valores > 1000)

5. O jeito 'lento' (loop com if)

```python
start = time.time()
vendas_altas_loop = []
for valor in faturamento_loop:
    if valor > 1000:
        vendas_altas_loop.append(valor)
end = time.time()
print(f"Tempo loop: {end - start:.5f} segundos")
```

6. O jeito 'rápido' 

```python
start = time.time()
vendas_altas_numpy = faturamento_numpy[faturamento_numpy > 1000]
end = time.time()
print(f"Tempo numpy: {end - start:.5f} segundos")
```

```python
import pandas as pd
```

2. lendo o arquivo 'vendas.csv' (trabalharemos em cima dele)

```python
file_path = r"C:\Users\matheus.r\Desktop\data-engineering-roadmap\week_02_python_dados\data\raw\vendas.csv"
dados = pd.read_csv(file_path)
print(dados)
```

3. função 'dropna'

```python
def limpeza_dados(dados):
    dados_limpos = dados.dropna()
    return dados_limpos

limpeza_dados(dados)
```

4. função 'fillna'

```python
def filtragem_dados(dados):
    dados[' valor'] = dados[' valor'].fillna(500.0)
    return dados

filtragem_dados(dados)
```

5. função 'astype'

```python
def tipo(dados):
    dados[' customer_id'] = dados[' customer_id'].astype(str)
print(dados.dtypes)
print(dados)
```

6. função 'rename'

```python
def renomear(dados):
    dados_novos = dados.rename(columns={' valor': ' preco'})
    return dados_novos
dados = renomear(dados)
print(dados.columns)
```

7. função 'apply'

```python
def classificar_venda(preco):
    if preco > 300:
        return 'Venda Alta'
    else:
        return 'Venda Normal'
dados['categoria'] = dados[' preco'].apply(classificar_venda)
print(dados)
```

8. função 'map'

```python
def remapear(dados):
    mapas_regionais = {
            101: ' Sudeste',
            102: ' Sul',
            103: ' Norte'
    }
    dados[' regionais'] = dados[' customer_id'].map(mapas_regionais)
    return dados

dados = remapear(dados)
print(dados.head())
```

9. criando e salvando o 'clean.csv'

```python
caminho_salvar = r"C:\Users\matheus.r\Desktop\data-engineering-roadmap\week_02_python_dados\data\processed\clean.csv"

os.makedirs('data/processed', exist_ok='True')
dados.to_csv(caminho_salvar, index='False')
```
